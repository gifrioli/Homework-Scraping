```python
from bs4 import BeautifulSoup
import requests

response = requests.get("https://www.lemonde.fr/")
doc = BeautifulSoup(response.text)
```

```python
items = doc.find_all('div', class_='article')
len(items)
```

```python
rows = []

for item in items:
    row = {}

    #Titles
    try:
        title_element = item.find(class_="article__title")
        row['title'] = title_element.text.strip()
    except:
        pass

    #Subheads
    try:
        subhead = item.find(class_="article__desc")
        row['subhead'] = subhead.text.strip()
    except:
        pass

    #URL
    try:
        url = item.find('a')
        if url and 'href' in url.attrs:
            row['url'] = url['href']
    except:
        pass

    #Imagens
    try:
        img = item.find('img')
        if img:
            if 'src' in img.attrs:
                row['image_url'] = img['src']
            elif 'data-src' in img.attrs:
                row['image_url'] = img['data-src']
    except: 
        pass

    #Premium
    try:
        premium = item.find(class_='icon__premium')
        if premium:
            row['premium'] = True
        else:
            row['premium'] = False
    except:
        row['premium'] = False
        
    #Byline
    try:
        byline = item.find(class_="meta__author meta__author--no-after")
        if not byline:
            byline = item.find(class_='article__author-link')
        row['byline'] = byline.text.strip()
    except: 
        pass

    #ArticleType
    try:
        article_type = item.find(class_='article__category')
        if not article_type:
            article_type = item.find(class_='category-link')
        row['article_type'] = article_type.text.strip()
    except:
        pass
    
    print(row)
    rows.append(row)
```

```python
import pandas as pd
df = pd.DataFrame(rows)
df.to_csv('lemonde_articles.csv', index=False)
```
