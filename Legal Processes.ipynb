```python
for page in range(1, 15):
    url = base_url.format(page)
    print(url)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    #I found two different blocks of results in the code.
    #Blocks: class=result_block and class=result_altblock

    result_blocks = soup.select('div.result_block, div.result_altblock')
    
    for block in result_blocks:
            row = {}
    
            #Cases Names
            #Class = "result_title"
            title = block.find('div', class_='result_title')
            if title:
                a_tag = title.find('a')
                if a_tag:
                    row['name'] = a_tag.text.strip()
                    row['url'] = a_tag['href']
    
                #Category
                #Class = "category"
                category_tag = title.find('span', class_='category')
                if category_tag:
                    row['category'] = category_tag.text.strip()
    
            #Info
            #Class = "infoline"
            info_tag = block.find('div', class_='infoline')
            if info_tag:
                row['info'] = info_tag.text.strip()
    
            all_cases.append(row)
    
    #Print data
    for case in all_cases:
        print(case)
```

```python
import pandas as pd

df = pd.DataFrame(all_cases)
df.to_csv('cases_cars.csv', index=False)
```
